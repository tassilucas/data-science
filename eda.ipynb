```python
import seaborn as sns

sns.set(rc = {'figure.figsize': (30, 30)})
sns.set(font_scale=4)
```

```python
from matplotlib import pyplot as plt
```

```python
def produce_boxplot_grid(dataset):
    sns.set(font_scale=1)
    fig, axes = plt.subplots(2, 3, figsize=(20, 20))
    sns.boxplot(ax=axes[0, 0], x = 'Sexo', y = 'Rend_mensal', data = dataset)
    sns.boxplot(ax=axes[0, 1], x = 'Anos_estudo', y = 'Rend_mensal', data = dataset)
    sns.boxplot(ax=axes[0, 2], x = 'uf', y = 'Rend_mensal', data = dataset)
    sns.boxplot(ax=axes[1, 0], x = 'PosOcup', y = 'Rend_mensal', data = dataset)
    sns.boxplot(ax=axes[1, 1], x = 'Instruc', y = 'Rend_mensal', data = dataset)
    sns.boxplot(ax=axes[1, 2], x = 'G_Anos_estudos', y = 'Rend_mensal', data = dataset)
    
    for ax in fig.axes:
        plt.sca(ax)
        plt.xticks(rotation=90)
```

```python
import scipy
```

```python
log_data = data.copy()
```

```python
log_data.Rend_mensal = np.log(log_data.Rend_mensal)
```

```python
log_data.Rend_mensal
```

```python
plt.scatter(x = log_data.Horas_trab, y = log_data.Rend_mensal)
plt.title('Gráfico de Dispersão entre Horas trabalhadas e Renda mensal')
plt.show()
```

Observando o gráfico de dispersão é possível perceber uma correlação positiva, onde a renda mensal aumenta conforme o aumento do número de horas, porém esta correlação é fraca (valores muito dispersos). Agora vamos encontrar os valores de covariância e coeficiente de Pearson e Spearman:

```python
log_data = log_data[log_data['Rend_mensal'].notna()]
```

```python
d = np.array([log_data.Horas_trab, log_data.Rend_mensal])
d
```

Covariância:

```python
np.cov(d[0], d[1])[0][0]
```

Coeficiente de Pearson:

```python
scipy.stats.pearsonr(d[0], d[1])
```

Coeficiente de Spearman:

```python
scipy.stats.spearmanr(d[0], d[1])
```

Com base nos valores obtidos pela covariância, percebe-se realmente uma correlação positiva. Além disso, pelos coeficientes de Pearson e Spearman, percebe-se uma fraca relação tanto linear quanto monotonica entre essas duas variavies. Sendo assim, não existe uma relação tão direta entre elas.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(d[0].reshape((-1, 1)), d[1])
```

```python
plt.plot(log_data.Horas_trab, log_data.Rend_mensal, 'o')
plt.plot(log_data.Horas_trab, model.coef_*log_data.Horas_trab + model.intercept_)
```

A constante corresponde ao local do eixo Y em que a reta cruza e o coeficiente angular representa a inclinação da reta de regressão, apresentando a quantidade em unidade que uma medida varia em função de outra.

Regressão linear robusta:

```python
from sklearn.linear_model import HuberRegressor

rob = HuberRegressor().fit(d[0].reshape((-1, 1)), d[1])
```

```python
plt.plot(log_data.Horas_trab, log_data.Rend_mensal, 'o')
plt.plot(log_data.Horas_trab, rob.coef_*log_data.Horas_trab + rob.intercept_)
```

```python
fig, ax1 = plt.subplots(1, 1)
fig.suptitle('Comparativo entre as duas retas de regressão geradas:')

ax1.plot(log_data.Horas_trab, log_data.Rend_mensal, 'o')
ax1.plot(log_data.Horas_trab, model.coef_*log_data.Horas_trab + model.intercept_, c='r')
ax1.plot(log_data.Horas_trab, rob.coef_*log_data.Horas_trab + rob.intercept_, c='g')

plt.tight_layout()
```

Visualmente é dificil perceber diferença significante entre as duas retas de regressão. No entanto, geralmente, retas robustas tendem a se comportar melhor na presença de outliers.

Tabela de contigência:

```python
F = pd.crosstab(data.G_Anos_estudos, data.Sexo)
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2, p_value, dof, expected = chi2_contingency(F)
```

```python
F
```

```python
expected
```

```python
def residuals(observed, expected):
    return (observed - expected) / np.sqrt(expected)

r = residuals(F, expected)
```
