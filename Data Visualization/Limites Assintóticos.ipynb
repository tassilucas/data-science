```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_demands(B, C):
    return [b/C for b in B]
```

Limites de __troughput__:

```python
def plot_troughput(N, interval, Z, demands, label, c):
    x = np.linspace(1, N, interval)
    d_max = max(demands)
    D = sum(demands)
    
    lower_load_limit = x/(D+Z)
    
    plt.axhline(y=1/d_max, linestyle='-', label=(label + " MAX"), color=c)
    plt.plot(x, lower_load_limit, label=label, color=c)
```

```python
busy_time = [400, 100, 600]
C = 200
Z = 15
```

```python
import matplotlib

matplotlib.rc('figure', figsize=(15, 10))
```

```python
# Sem mudanas
demands = find_demands(busy_time, C)
plot_troughput(40, 8, Z, demands, "None", "r")

# 1
plot_troughput(40, 8, Z, [1, 0.5, 3.0], "CPU 2 vezes mais rapido", "c")

# 2
plot_troughput(40, 8, Z, [2, 2.06, 2.06], "Balanceado", "m")

# 3
plot_troughput(40, 8, Z, [2, 0.5, 1.5, 1.5], "Disco mais rapido", "k")

# 4
plot_troughput(40, 8, Z, [1, 1.27, 1.27, 1.27], "Todas as 3", "b")

plt.xlabel('N')
plt.ylabel('X(N)')
plt.legend()
plt.show()
```

Limites de __tempo de resposta__:

```python
def plot_response(N, interval, Z, demands, label, c):
    x = np.linspace(1, N, interval)
    d_max = max(demands)
    D = sum(demands)
    
    limit = x*d_max - Z
    
    plt.axhline(y=D, linestyle='-', label=(label + " MIN"), color=c)
    plt.plot(x, limit, label=label, color=c)
```

```python
# Sem mudanas
plot_response(40, 8, 15, demands, "None", "r")

# 1
plot_response(40, 8, 15, [1, 0.5, 3.0], "CPU 2 vezes mais rapido", "c")

# 2
plot_response(40, 8, Z, [2, 2.06, 2.06], "Balanceado", "m")

# 3
plot_response(40, 8, Z, [2, 0.5, 1.5, 1.5], "Disco mais rapido", "k")

# 4
plot_response(40, 8, Z, [1, 1.27, 1.27, 1.27], "Todas as 3", "b")

plt.xlabel('N')
plt.ylabel('R(N)')
plt.ylim(0, 30)
plt.xticks(np.arange(0, 41, 8, dtype=int)) 
plt.legend()
plt.show()
```
