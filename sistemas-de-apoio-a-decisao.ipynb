```python
import musicbrainzngs

musicbrainzngs.auth("tassilucas", "Matheu$Donato18")

# Tell musicbrainz what your app is, and how to contact you
# (this step is required, as per the webservice access rules
# at http://wiki.musicbrainz.org/XML_Web_Service/Rate_Limiting )
musicbrainzngs.set_useragent("Get genre from artists", "0.1", "na")
```

```python
def get_genre(row):
    artist_name = row['artist']
    result = musicbrainzngs.search_artists(artist=artist_name)['artist-list'][0]['tag-list']
    list_of_genres = sorted(result, key=lambda d: int(d['count']), reverse=True)
    print("{}: {}".format(artist_name, list_of_genres[0]['name']))
    return str(list_of_genres[0]['name'])

valuable_artists['tag'] = valuable_artists.apply(lambda x: get_genre(x), axis=1)
```

```python
valuable_artists['tag'].value_counts().plot.pie(figsize=(10, 20))
```
